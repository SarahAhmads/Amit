```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    return n * fact(n-1)
fact(5)
```

```python
def is_prime(n):
    if n == 1:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
        return True
is_prime(1)
```

```python
def reverse(s):
    split_s = s.split()
    new_list = split_s[::-1]
    reversed_s = " ".join(new_list)
    return reversed_s
reverse("Hello World")
```

```python
import random
def generate_password(length):
    """
    this function generates a random password
    :param_1 length : takes the length of the password from input value
    type param_1 length : int
    return : this function will return a randomly generated password
    rtype : str
    """
    characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()"
    password = ""
    for _ in range(length):
        password += random.choice(characters)
    return password

print(f"Generated Password: " + generate_password(12))
```

```python
from abc import ABC, abstractmethod
class Animal(ABC):

    @abstractmethod
    def move(self):
        pass
class Dog(Animal):
    def move(self):
        return "Dog walks on 4 legs"
    
class Bird(Animal):
    def move(self):
        return "Bird flies in the sky"
    
class Fish(Animal):
    def move(self):
        return "Fish swims in water"

dog = Dog()
print(dog.move())
```

```python
import random
def generate_password(length):
    """
    This function generates a random password.
    :param length: The desired length of the password.
    :type length: int
    :return: A randomly generated password.
    :rtype: str
    """
    characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*()"
    password = ""
    for _ in range(length):
        password += random.choice(characters)
    return password
```
