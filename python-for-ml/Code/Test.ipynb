```python
Grade = int(input("Enter your Grade: "))
if Grade >= 0 and Grade <= 100:
    if 100 >= Grade >= 90:
        print("A")
    elif 90 >= Grade >= 80:
        print("B")
    elif 80 >= Grade >= 70:
        print("C")
    elif 70 >= Grade >= 60:
        print("D")
    else:
        print("F")
else:
    print("Please enter a value between 0 and 100")
```
